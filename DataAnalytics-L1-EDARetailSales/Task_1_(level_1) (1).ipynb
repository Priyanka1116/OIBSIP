```python
df['Date'] = pd.to_datetime(df['Date'])     #### Observation

#The dataset contains both categorical and numerical variables. The Date column was initially stored as an object and needs conversion to datetime format for time series analysis.
```

```python
df.dtypes
```

```python
df.describe() #it decribe thee over data in number
```

```python
print("Mean Sales:", df['Sales'].mean())

print("Median Sales:", df['Sales'].median())

print("Mode Sales:")
print(df['Sales'].mode())

print("Standard Deviation:", df['Sales'].std())

#here we are doing sales calucation in all type
```

```python
df['Month'] = df['Date'].dt.month
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum()

monthly_sales
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

monthly_sales.plot(marker='o',color='purple')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.grid(True)

plt.show()
#### Observation

#Sales were highest in Month X and lowest in Month Y. The business experiences fluctuations in monthly revenue.
```

```python
df['Quarter'] = df['Date'].dt.quarter
df['Quarter']
```

```python
quarter_sales = df.groupby('Quarter')['Sales'].sum()
```

```python
plt.figure(figsize=(6,4))

quarter_sales.plot(marker='o')

plt.title('Quarterly Sales Trend')
plt.xlabel('Quarter')
plt.ylabel('Sales')

plt.grid(True)

plt.show()
#### Observation #Quarter 3 generated the highest sales revenue, indicating stronger customer demand during that period.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))

sns.countplot(x='Gender', data=df,color='lightgreen')

plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Number of Customers')

plt.show()
### Observation

#The customer base is almost equally distributed between male and female customers.
```

```python
product_sales = df.groupby('Product line')['Sales'].sum()

product_sales = product_sales.sort_values(ascending=False)

plt.figure(figsize=(10,5))

product_sales.plot(kind='bar',color='orange')

plt.title('Revenue by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')

plt.show()
### Observation

#Food and beverages generated the highest sales revenue among all product lines.
```

```python
top_products = df.groupby('Product line')['Quantity'].sum()

top_products = top_products.sort_values(ascending=False)

plt.figure(figsize=(8,5))

top_products.plot(kind='bar',color='pink')

plt.title('Most Sold Product Lines')
plt.xlabel('Product Line')
plt.ylabel('Quantity Sold')

plt.show()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(x='Payment', data=df)

plt.title('Payment Method Distribution')

plt.show()
### Observation

#Electronic payment methods are preferred by most customers.
```

```python
plt.figure(figsize=(7,4))

sns.histplot(df['Rating'], bins=10,color='red')

plt.title('Customer Ratings')

plt.xlabel('Rating')

plt.show()
```

```python
numeric_columns = df.select_dtypes(include=['int64', 'float64'])
```

```python
correlation = numeric_columns.corr()

correlation
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm'
)

plt.title("Correlation Heatmap")

plt.show()


### Observation #sales and gross income have a strong positive correlation. Higher sales contribute to increased gross income.
```

```python
avg_rating = df.groupby('Product line')['Rating'].mean()

plt.figure(figsize=(8,5))

avg_rating.plot(kind='bar',color='coral')

plt.title('Average Customer Rating by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Average Rating')

plt.show()


### Observation

#Certain product lines receive higher customer ratings, indicating better customer satisfaction.
```
